Now we separate data into two groups, one group is the occurances of a certain category hurricane before year 2000, and another group after year 2000. One often used statistical method to find out whether there is a difference between the two groups is ANOVA test.

ANOVA test:

Analysis of variance (ANOVA) is a statistical test used to evaluate the difference between the means of two or more groups

```python
from IPython.display import Image
Image(filename='anova2.jpg') 
```

Assumption of anova test is that the data has to be normally distributed.

The following function given the hurricane category, and separate number of hurricanes for each year into two groups, one before 2000, and one after 2000. It then performs normal test on the two groups.

```python
from scipy.stats import f_oneway
from scipy import stats
from scipy.stats import mannwhitneyu



def normalTest(category):

    dfcat = dfNew['category'] == category
    dfTest = dfNew[dfcat]
    
    beginYear = 1975
    endYear = 2021
    lst = []
    for i in range(endYear - beginYear + 1):
        lst.append(0)

    for year in dfTest['year']:
        lst[year-beginYear] += 1
    
    earlierLst = lst[:2000-beginYear]
    laterLst = lst[2000-beginYear:]
    print(earlierLst)
    print(laterLst)
    resEarly = stats.normaltest(np.array(earlierLst))
    print("p value for earlier group: ", resEarly.pvalue)
    resLater = stats.normaltest(np.array(laterLst))
    print("p value for later group: ", resLater.pvalue)

```

The following function perform Mann Witney U test on a certain category of hurricanes, for groups before and after 2000

```python
def mann_whitney_u_test(category):

    dfcat = dfNew['category'] == category
    dfTest = dfNew[dfcat]
    
    beginYear = 1975
    endYear = 2021
    lst = []
    for i in range(endYear - beginYear + 1):
        lst.append(0)

    for year in dfTest['year']:
        lst[year-beginYear] += 1
    
    earlierLst = lst[:2000-beginYear]
    laterLst = lst[2000-beginYear:]
    
    U1, p = mannwhitneyu(earlierLst, laterLst, method="exact")
    print('p-value for mann whitney u test for category', category, 'is: ', p)
   
```

```python
for i in [5,4,3]:
    mann_whitney_u_test(i)
```

For category 5 hurricanes, there is no difference before and after 2000.
For category 4 hurricanes, the chance they are equal is 9.0%
For category 3 hurricanes, the chance they are equal is 5.3%

Statisticians like to use 5% as the cut off point for any definitive answer whether the difference is statistically significant, so from the data we have, the answer is not definitive.

But let's think about another question:  is there a trend in the occurance of hurricanes? Is it increasing, or decreasing?

There is a trend test in numpy that would find a trend in a list of numbers, whether the trend is increasing, decreasing, or no trend. The idea is to draw a best fit line among the points (number of occurances of a hurricanes in a particular year is a point).

```python
Image(filename='best-fit.png') 
```

If the line turns upwards, the trend is increasing. If the line turns downwards, the trend is decreasing.

The following function would fit a best-fit line among "points" of occurances of a particular hrricane category, and returns the slope of the line. If the slope is positive, the line turns upwards. If the slope is negative, the line turns downwards. It also draws the best fit line

```python
def trendline(index,data, order=1):
    coeffs = np.polyfit(index, list(data), order)
    slope = coeffs[0]
    return float(slope), coeffs

def trendTest(category):

    dfcat = dfNew['category'] == category
    dfTest = dfNew[dfcat]
    
    beginYear = 1975
    endYear = 2021
    lst = []
    index = []
    for i in range(endYear - beginYear + 1):
        lst.append(0)
        index.append(i+1975)

    for year in dfTest['year']:
        lst[year-beginYear] += 1

    resultent, trend =trendline(index,lst)

    print('For category', category, 'hurricanes, the slope is:',resultent)  

    # plot the scatter points
    plt.plot(index, lst, 'o')

    # calculate the trendline
    trendpoly = np.poly1d(trend) 

    # plot the trend line
    plt.plot(index, trendpoly(index))

    plt.title('Trend line for category ' + str(category) + ' hurricanes')
    plt.show()
```

```python
for i in [5, 4, 3]:
    trendTest(i)
```

Tests shows positive slopes for all categories 3, 4, and 5 hurricanes, which means an increasing trend.
